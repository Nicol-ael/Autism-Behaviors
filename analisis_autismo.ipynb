```python
# Carga del dataset principal desde archivo CSV
# Este paso permite incorporar los datos crudos al entorno de análisis
# y constituye el punto de partida del proceso de preparación (ETL)

df = pd.read_csv("Toddler Autism dataset July 2018.csv")



```

</h2>

```python
# Inspección inicial del dataset
# Se visualizan las primeras filas para validar la estructura de las variables,
# identificar posibles inconsistencias y comprender la organización general de los datos
# antes de iniciar el proceso de limpieza y transformación

df.head(10)
```

```python
# Análisis de la estructura del dataset
# Se evalúan tipos de datos, cantidad de registros y presencia de valores nulos
# con el objetivo de identificar posibles problemas de calidad
# que deban abordarse en la fase de limpieza

df.info()




```

```python
# Evaluación de la dimensión del dataset
# Se identifican el número de registros y variables para dimensionar el volumen de información
# y entender el alcance del análisis previo a la transformación

df.shape
```

```python
# Estandarización de los nombres de las variables
# Se normaliza el formato eliminando espacios, unificando a minúsculas y reemplazando "_" por "espacios"
# con el objetivo de garantizar consistencia en la nomenclatura y evitar errores en el manejo del dataset

df.columns = df.columns.str.strip().str.lower().str.replace("_", " ")

```

```python
# Estandarización de variables categóricas
# Se seleccionan las columnas de tipo texto para aplicar limpieza únicamente sobre estas variables
# y normalizar su formato, evitando inconsistencias en los valores (espacios, mayúsculas/minúsculas)


text_cols = df.select_dtypes(include=['object', 'string']).columns
df[text_cols] = df[text_cols].apply(lambda x: x.str.strip().str.lower())



```

```python
# Formateo de nombres de variables para presentación
# Se aplica un tratamiento diferenciado: las variables tipo A1, A2, etc. se mantienen en mayúscula
# para preservar su estructura original, mientras que el resto se transforma a un formato más legible
# (reemplazo de "_" por espacios y uso de formato título)



df.columns = [
    col.upper() if col.lower().startswith('a') and col[1:].isdigit()
    else col.replace('_', ' ').title()
    for col in df.columns]
```

```python
# Eliminación de registros duplicados
# Se asegura la autonomia de cada observación dentro del dataset



df = df.drop_duplicates()

```

```python
# Evaluación de valores faltantes por variable
# Se cuantifican los valores nulos en cada columna para identificar problemas de completitud



df.isnull().sum()
```

```python
# Validación posterior a la limpieza
# Se visualizan las primeras filas para verificar la correcta aplicación
# de las transformaciones realizadas y confirmar la consistencia del dataset

df.head()
```

```python
# Estandarización de la variable categórica 'sex'
# Se unifica el formato de los valores convirtiéndolos a mayúsculas

df['Sex'] = df['Sex'].str.upper()
```

```python
# Transformación de la variable de edad
# Se genera una nueva variable en años completos a partir de la información original
# con el objetivo de facilitar su interpretación y permitir un análisis más claro por grupos etarios

df['Age Years'] = (df['Age Mons'] // 12).astype(int)

```

```python
# Eliminación de la variable original 'age_mons'
# Una vez transformada la edad a años completos, esta variable resulta redundante
# y se elimina para mantener un dataset más limpio y evitar duplicidad de información

df = df.drop(columns=['Age Mons'])
```

```python
# Eliminamos la columna 'Qchat-10-Score' y 'Age Mons'
# El dataset recomienda excluirla para evitar posibles sesgos o riesgos en el análisis
# Esto asegura un modelo más limpio y confiable

df = df.drop(columns=['Qchat-10-Score'])
```

```python
# Guardamos los nombres de las columnas en una lista
# para poder reorganizar el orden del dataset de forma más clara
cols = list(df.columns)

```

```python
# Quitamos la columna 'age_years' de su posición actual
# para poder ubicarla después en otra parte del dataset
cols.remove('Age Years')
```

```python
# Definimos la posición donde queremos insertar la variable 'Age Years' despues de la variable 'sex'
idx = cols.index('Sex')
```

```python

# Insertamos la columna en la nueva posición dentro del dataset
# para mantener un orden más lógico de las variables
cols.insert(idx + 1, 'Age Years')
```

```python
# Aplicamos el nuevo orden de columnas al dataset
# reorganizando las variables según la estructura definida previamente
df = df[cols]
```

```python
# Visualizamos las primeras filas del dataset para verificar
# que el reordenamiento de columnas se aplicó correctamente
df.head()
```

```python
# Estandarizamos la variable 'ethnicity'
# Unificamos el formato eliminando espacios y normalizando a minúsculas
# para evitar inconsistencias en las categorías antes de su análisis
df['Ethnicity'] = df['Ethnicity'].str.strip().str.lower()
```

```python
# Definición de mapeo para la recategorización de la variable 'ethnicity'
# Se agrupan categorías similares para reducir la fragmentación de los datos
# y facilitar su análisis e interpretación en el dashboard

mapping = {
    'asian': 'asian',
    'south asian': 'asian',
    
    'white european': 'white',
    
    'middle eastern': 'middle.east',
    
    'hispanic': 'latino',
    'latino': 'latino',
    
    'black': 'black',
    
    'native indian': 'other',
    'pacifica': 'other',
    'mixed': 'other',
    'others': 'other'
}
```

```python
# Aplicación del mapeo a la variable 'ethnicity'
# Se recategorizan los valores según el diccionario definido y
# aquellos no contemplados se asignan a 'other' para evitar valores nulos
# y mantener consistencia en las categorías
df['Ethnicity'] = df['Ethnicity'].map(mapping).fillna('other')
```

```python

# Validación de categorías en la variable 'ethnicity'
# Se revisan los valores únicos para confirmar que la recategorización
# se aplicó correctamente y que no existen inconsistencias en las categorías
df['Ethnicity'].unique()
```

```python
# Renombramos las variables A1–A10 a nombres descriptivos
# para hacer más clara la interpretación de cada comportamiento evaluado
# y facilitar su análisis y visualización en el dashboard

df = df.rename(columns={
    'A1': 'Response_To_Name',
    'A2': 'Eye_Contact',
    'A3': 'Pointing_To_Request',
    'A4': 'Pointing_To_Share',
    'A5': 'Pretend_Play',
    'A6': 'Following_Gaze',
    'A7': 'Empathy',
    'A8': 'Language',
    'A9': 'Gestures',
    'A10': 'Attentive_eye_c'
})
```

```python
# Definimos una lista con las variables de comportamiento
# Estas variables representan los rasgos evaluados en el test
# y se utilizarán para calcular la puntuación total de traits de ASD
cols = [
    'Response_To_Name','Eye_Contact','Pointing_To_Request',
    'Pointing_To_Share','Pretend_Play','Following_Gaze',
    'Empathy','Language','Gestures','Attentive_eye_c'
]
# Visualizamos las primeras filas solo de estas columnas
# Esto facilita enfocarnos en los comportamientos y validar sus valores
df[cols].head()
```

```python
# Definimos una lista con las variables de comportamiento
# Estas variables corresponden a los indicadores evaluados en el test
# y se utilizarán para analizar y calcular los rasgos asociados al ASD
cols = [
    'Response_To_Name','Eye_Contact','Pointing_To_Request',
    'Pointing_To_Share','Pretend_Play','Following_Gaze',
    'Empathy','Language','Gestures','Attentive_eye_c'
]
```

```python
# Cálculo de la puntuación total de rasgos por individuo
# Se suman las variables de comportamiento para obtener un score agregado
# que representa la intensidad de rasgos asociados al ASD en cada caso

df['Total_Traits'] = df[cols].sum(axis=1)
```

```python
# Clasificación del puntaje total de rasgos
# Se segmenta el score en niveles para facilitar su interpretación
df.Total_Traits.head()
```

```python
# Se creo una variable categórica basada en el total de rasgos
# y permitir el análisis por categorías de intensidad (low, medium, high)

df['Trait_Level'] = pd.cut(
    df['Total_Traits'],
    bins=[-1,3,6,10],
    labels=['Low','Medium','High']
)
```

```python
# Visualizamos los primeros registros de la nueva variable 'Trait Level'
# Esto nos permite verificar cómo quedó la clasificación en niveles
# Validamos que la categorización se haya aplicado correctamente

df ['Trait_Level'].head(50)
```

```python
# Análisis de la relación entre el score de rasgos y la clasificación de ASD
# Se comparan ambas variables para evaluar la coherencia entre el puntaje calculado
# y la variable de diagnóstico, identificando posibles patrones en los datos

df[['Total_Traits', 'Class/Asd Traits']].head(10)
```

```python
# Agrupamos los datos por nivel de rasgos (Trait_Level)
# Contamos la frecuencia de cada categoría de autismo dentro de cada nivel
# Esto nos permite analizar cómo se distribuye el diagnóstico según la intensidad de rasgos 
df.groupby('Trait_Level')['Class/Asd Traits'].value_counts()
```

```python
# Contamos la frecuencia de cada categoría en la variable de diagnóstico
# Esto nos permite ver cuántos casos son "yes" y cuántos "no"
# Es útil para entender la distribución inicial del dataset

df['Class/Asd Traits'].value_counts()
```

```python
# Convertimos la variable de diagnóstico a formato numérico
# Asignamos 1 a "yes" y 0 a "no" para facilitar el análisis
# Esto permite usar la variable en cálculos y modelos

# Contamos la cantidad de casos en cada categoría
# Nos ayuda a entender la distribución del diagnóstico en el dataset

df['ASD'] = df['Class/Asd Traits'].map({'yes':1, 'no':0})
print(df['ASD'].value_counts())
```

```python
# Configuramos pandas para mostrar todas las columnas del DataFrame
# Esto evita que se oculten columnas cuando el dataset es muy ancho
# Facilita la exploración completa de los datos 

pd.set_option('display.max_columns', None)
df.head()
```

```python
# Revisamos el tipo de dato de cada columna
# Esto nos permite entender cómo está estructurada la información
# Es clave para validar que cada variable tenga el formato correcto para el análisis

df.dtypes
```

```python
# Seleccionamos un subconjunto específico de columnas del dataset
# Incluimos el identificador y las variables de comportamiento relevantes
# Usamos .copy() para trabajar sobre una copia y evitar modificar el dataset original

facts_behaviors= df[[
    'Case No',
    'Response_To_Name',
    'Eye_Contact',
    'Pointing_To_Request',
    'Pointing_To_Share',
    'Pretend_Play',
    'Following_Gaze',
    'Empathy',
    'Language',
    'Gestures',
    'Attentive_eye_c',
   
]].copy()
```

```python
# Definimos una lista con las variables de comportamiento
# Estas columnas representan los indicadores clave del análisis
# Nos permitirá trabajar de forma organizada y enfocada en los rasgos

cols_comportamientos = [
    'Response_To_Name',
    'Eye_Contact',
    'Pointing_To_Request',
    'Pointing_To_Share',
    'Pretend_Play',
    'Following_Gaze',
    'Empathy',
    'Language',
    'Gestures',
    'Attentive_eye_c'
]
```

```python
# Definimos la columna identificadora del dataset
# 'Case No' funciona como ID único para cada registro
# Esto nos permite rastrear y organizar cada observación correctamente

id_col = 'Case No'
```

```python
# Transformamos el dataset de formato ancho a formato largo
# Convertimos múltiples columnas de comportamiento en dos columnas: tipo de comportamiento y valor
# Esto facilita el análisis, visualización y comparación entre variables

df_long = df.melt(
    id_vars=['Case No'],
    value_vars=[
        'Response_To_Name',
        'Eye_Contact',
        'Pointing_To_Request',
        'Pointing_To_Share',
        'Pretend_Play',
        'Following_Gaze',
        'Empathy',
        'Language',
        'Gestures',
        'Attentive_eye_c'
    ],
    var_name='Comportamiento',
    value_name='Valor'
)
```

```python
# Contamos cuántas veces aparece cada ID en el dataset transformado
# Esto nos permite verificar que cada caso se repite correctamente por cada comportamiento
# Validamos que la transformación a formato largo se hizo bien

df_long['Case No'].value_counts().head()
```

```python
# Transformamos la tabla de comportamientos a formato largo
# Convertimos las columnas de comportamientos en filas con su respectivo valor
# Esto permite analizar cada comportamiento de forma individual por caso

facts_behaviors = facts_behaviors.melt(id_vars=['Case No'], value_vars=cols_comportamientos, var_name='Comportamiento', value_name='Valor')
```
